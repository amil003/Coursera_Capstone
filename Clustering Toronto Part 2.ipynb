```python
import pandas as pd

toronto_df=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
toronto_df_drop = toronto_df[toronto_df.Borough != "Not assigned"].reset_index(drop=True)
for index, row in toronto_df_drop.iterrows():
    if row["Neighbourhood"] == "Not assigned":
        row["Neighbourhood"] = row["Borough"]
        
toronto_df_drop
```

### Then let's read the csv file to create a dataframe showing the latitudes and longitudes.

```python
csv_path = 'http://cocl.us/Geospatial_data'

coordinates = pd.read_csv(csv_path)
coordinates.head()
```

### Now that we have the latitude and longitude of each location, let's go ahead and merge the two dataframes.

```python
toronto_df_merge = toronto_df_drop.merge(coordinates, on="Postal Code", how="left")
toronto_df_merge.head()
```

### And let's make it match the table on coursera.

```python
column_names = ["Postal Code", "Borough", "Neighbourhood", "Latitude", "Longitude"]
test_df = pd.DataFrame(columns=column_names)

test_list = ["M5G", "M2H", "M4B", "M1J", "M4G", "M4M", "M1R", "M9V", "M9L", "M5V", "M1B", "M5A"]

for postcode in test_list:
    test_df = test_df.append(toronto_df_merge[toronto_df_merge["Postal Code"]==postcode], ignore_index=True)
    
test_df
```
